```python
import pandas as pd
from pandas import *
import numpy as np
```

```python
def riskrewardpct(risk,reward):
    return float(risk)/float(risk+reward)

riskreward = {}

pos = 'BB'
# Risk
for i in np.arange(3,20,.1):
    riskreward[i] = []
    # Reward
    for j in np.arange(3.5,15,.1):
        riskreward[i].append(riskrewardpct(i,j))

df = DataFrame.from_dict(riskreward)
df.set_index(np.arange(3.5,15,.1), inplace = True)
df.index.name = 'Reward'
df.columns.name = 'Risk'

df
```
